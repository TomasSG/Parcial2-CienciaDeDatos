# Preparación de datos

En esta sección comenzamos definiendo las funciones necesarias para aplicar la preparación de datos. Luego, generaremos diferentes datasets para realizar el entrenamiento de modelos y encontrar aquellos que tengan una mejor perfomance.

## Definición funciones
Con el objetivo de modularizar la función preparacion_datos, vamos a constuir algunas funciones que le sirven de soporte.

````python
# FUNCION imputar_nulos
# Para imputar los nulos en Arrival Delay in Minutes se decide reemplazar los valores por la media.
def imputar_nulos(df):

  # Copiar el df para no modificarlo in situ
  df_c = df.copy()

  # Calcular la media de las observaciones, sin tener en cuenta los nulos
  index = pd.notnull(df_c["Arrival Delay in Minutes"])
  media = df_c[index]["Arrival Delay in Minutes"].mean()
  
  # Reemplazar los valores
  df_c["Arrival Delay in Minutes"] = df_c["Arrival Delay in Minutes"].fillna(media)

  return df_c
````

````python
# FUNCION calcular_media_por_derecha
# Calcula la media de la variable indicada, sin tener encuenta los valores que sean mayores al maximo dado
def calcular_media_por_derecha(df, var, max):
  
  # Filtrar los datos mayores a max
  index = df[var] <= max
  df_aux = df[index]

  # Calcular la media con los datos filtrados
  media = df_aux[var].mean()

  return media

# FUNCION imputar_outliers
# Para imputar los outliers en las variables necesarias, siempre por derecha.
def imputar_outliers(df):

  # Copiar el df para no modificarlo in situ
  df_c = df.copy()

  # Definir variables a las que aplicar la imputacion
  vars = ["Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]

  for var in vars:

    # Definimos cual será el outlier superior
    outlier_sup = df_c[var].mean() + 2 * df_c[var].std()

    # Calculamos la media sin valores mayores al outlier
    media_sin_outlier = calcular_media_por_derecha(df_c, var, outlier_sup)

    # Imputamos los valores
    df_c[var] = [x if x <= outlier_sup else media_sin_outlier for x in df_c[var]]

  return df_c
````

````python
# FUNCION crear_variables
# Para aplicar feature engineering.
def crear_variables(df):

  # Copiar el df para no modificarlo in situ
  df_c = df.copy()

  # 1) Variable que representa el delay total en un viaje
  df_c["Total Delay in Minutes"] = df_c["Departure Delay in Minutes"] + df_c["Arrival Delay in Minutes"]

  # 2) Diferencia entre los delay, para conocer cómo varía
  df_c["Diff Delay in Minutes DA"] = df_c["Departure Delay in Minutes"] - df_c["Arrival Delay in Minutes"]

  return df_c
````

````python
# FUNCION preparacion_datos
# df                ->  dataset de entrada a modifiar
# flag_outliers  ->  flag para imputar outliers
# flag_nulos     ->  flag para imputar nulos
# flag_vars        ->  flag para feature engineering
def preparacion_datos(df, flag_outliers, flag_nulos, flag_vars):

  # Crear una copia
  df_c = df.copy()

  # ---- IMPUTACIÓN DE OUTLIERS ----
  if flag_outliers:
    df_c = imputar_outliers(df_c)

  # ---- IMPUTACIÓN NULOS ----
  if flag_nulos:
    df_c = imputar_nulos(df_c)

  # ---- FEATURE ENGINEERING ----
  if flag_vars:
    df_c = crear_variables(df_c)

  # ---- TARGET ----
  # Converitr target en numerica
  # 0 ->  neutral or dissatisfied 
  # 1 ->  satisfied
  df_c["CLASE"] = [1 if x == "satisfied" else 0 for x in df_c["CLASE"]]
  y = np.array(df_c["CLASE"])

  # ---- FEATURES ----
  X = df_c.drop("CLASE", axis=1)

  # Normalizar las numéricas
  num_cols = X.columns[X.dtypes.apply(lambda c: np.issubdtype(c, np.number))]
  scaler_X = StandardScaler(with_mean=True, with_std=True)
  X[num_cols] = scaler_X.fit_transform(X[num_cols])

  # Obtenemos las dummies
  X = pd.get_dummies(X)

  return y, X
````

Todos se construyen aplicando una optimización de hiperparámetros, usando un mínimo de **15** valores para cada hiperparámetro.

Se supone que a la empresa le interesa detectar mayor cantidad de positivos, y que cada postivo tiene un valor del doble que el de un negativo. Entonces, se define una función de ganancia custom para la implementación, según los objetivos de negocio.

## Función Ganancia Custom

Como se menciona antes, **se trabaja sobre la suposición de que cada positivo vale el doble que los negativos**. Se averigua cuál sera la probabilidad de corte. Tener en cuenta que:

* #feliz = cantidad de observaciones cuya clase es satisfecho.
* #infeliz = cantidad de observaciones cuya clase es neutral o insatisfecho.


```
Ganancia = 2 * #feliz - 1 * #infeliz
```

Como se busca que la gananica sea mayor a cero



```
Ganancia > 0
2 * #feliz - 1 * #infeliz > 0
2 * #feliz + #feliz - #feliz - #infeliz > 0
3 * #feliz - (#feliz + #infeliz) > 0
3 * #feliz  > (#feliz + #infeliz)
#feliz / (#feliz + #infeliz) > 1/3
P(feliz) > 1/3
```
Teóricamenta se calcula que la probabilidad de corte debe ser de por lo menos un 1/3, para así favorecer a la detección de positivos, ya que tienen mayor preponderancia para el negocio



````python
#Definimos la función de ganancia a utilizar para la optimización
def funcion_ganancia(model, X, y_test):

    # Definir la probabilidad de corte calculada
    p_corte = float(1/3)

    # Calcular las probabilidades con el modelo
    y_pred = model.predict_proba(X)[:, 1]
    
    # Calcular la ganancia
    ganancia = sum([(2 if y_test[i] > 0 else -1) 
                    if y_pred[i]  > p_corte
                    else 0 
                    for i in range(len(y_pred))])
    
    return ganancia
````

## Matriz de Confusión

Aquí analizamos la matriz de confusió, y calculamos las métricas derivadas de esta. En resumen, la matriz de confusión consta de una matriz de 4x4 tal que:

Captura de pantalla 2021-07-12 163701.jpg

Donde:

* Accuracy: (Vp + Vn) / (Vp + Vn + Fp + Fn)
* Precision: Vp / (Vp + Fp)
* Sensibilidad: Vp / (Vp + Fn)
* Especificidad: Vn / (Fp + Vn)


Vamos a usar una función para ello.



````python

def matriz_confusion_metricas(model, X_test, y_test, nombre_model):

  # Inicializamos las listas con los resultados
  nombres = [
             "Accuracy",
             "Precision",
             "Sensibilidad",
             "Especificidad"
             ]
  valores = []

  # Calculamos las prediccioens (clases)
  y_pred = model.predict(X_test)

  # Creamos la matriz de confusión
  conf = confusion_matrix(y_test, y_pred)

  # Vamos a crear un dataframe para que sea más estético
  nombre_niveles = ["0", "1"]
  pred_cols = ["Pred_" + str(c) for c in nombre_niveles]
  df_conf = pd.DataFrame(conf, index=nombre_niveles, columns=pred_cols)

  # Valores elementales
  vp = df_conf["Pred_1"][1]
  fn = df_conf["Pred_0"][1]
  fp = df_conf["Pred_1"][0]
  vn = df_conf["Pred_0"][0]
  
  # Calculamos las métricas
  accuracy = (vp + vn) / (vp + vn + fp + fn)
  precision = vp / (vp + fp)
  sensibilidad = vp / (vp + fn)
  especificidad = vn / (fp + vn)
  
  # Agregar los resultados
  valores.append(round(accuracy, 4))
  valores.append(round(precision, 4))
  valores.append(round(sensibilidad, 4))
  valores.append(round(especificidad, 4))

  # Crear df con metricas
  metricas = pd.DataFrame({
      "Nombre": nombres,
      "Valor": valores
  })

  # Mostramos por pantalla resultado
  print("<" + "-" * 100 + ">")
  print("* MODELO {} *".format(nombre_model))
  print("\n")
  print("MATRIZ CONFUSIÓN")
  print("\n")
  print(df_conf)
  print("\n")
  print("MÉTRICAS")
  print("\n")
  print(metricas)
  print("\n")

  return accuracy, precision, sensibilidad, especificidad
````

Guardamos los resultados para usarlos en la conclusión final

````python
# Inicializamos las lista donde guardamos los resultados
accuracys = []
precisions = []
sensibilidades = []
especificidades = []

for i in range(df_modelos.shape[0]):

  # Obtener los datos importantes
  model = df_modelos["Modelo"][i]
  X_test = df_modelos["X_test"][i]
  y_test = df_modelos["y_test"][i]
  nombre_model = df_modelos["Nombre"][i]

  # Generar matriz de confusión
  accuracy, precision, sensibilidad, especificidad = matriz_confusion_metricas(model, X_test, y_test, nombre_model)

  # Guardamos los resultados
  accuracys.append(accuracy)
  precisions.append(precision)
  sensibilidades.append(sensibilidad)
  especificidades.append(especificidad)

# Guardamos los resultados en el df
df_modelos["Accuracy"] = accuracys
df_modelos["Precision"] = precisions
df_modelos["Sensibilidad"] = sensibilidades
df_modelos["Especificidad"] = especificidades
````

# Clustering

Buscamos agrupar a clientes según algunos factores demográficos, esto permite armar arquetipos de clientes y así generar mejores estrategias.

Para esta tareas vamos a hacer uso del algoritmo **K-means**.

Vamos a hacer uso de los datos escalados, con las variables categóricas encodeadas de alguna forma, con nulos imputados pero sin imputar outliers porque pueden aportar información valiosa.

No la vamos a aplicar con todas las variables del dataset, sino solo que aquellas que nos interesan para describir a los consumidores. Por ejemplo, si interesan las edades, géneros y distancias para segmentar.

````python
# Variables para el análisis
vars = ["Age", 
        "Flight Distance", 
        "Gender_Female", 
        "Gender_Male", 
        "Customer Type_Loyal Customer",
        "Customer Type_disloyal Customer",
        "Type of Travel_Business travel",
        "Type of Travel_Personal Travel",
        "Class_Business",
        "Class_Eco",
        "Class_Eco Plus"]

# Los datos a usar
X = X_con_outliers[vars]
````

En principio no se conoce el K (cantidad de clusters) ideal, por lo que vamos a utilizar métricas de validación interna:

* **Cohesión**: mide la distancia entre los puntos internos de cada cluster.
* **Separación**: mide la separación entre distintos clusters.

En base a esto buscamos los clusters

````python
# Defninir el rango de  K a probar
k = range(3, 30, 1)

# Función soporte para calcular el score para cada k
def entrenar_kmeans(df, k):

  # Definir el objeto
  k_means = KMeans(n_clusters=k, n_init=10, n_jobs=N_JOBS)

  # Entrenar el modelo
  k_means.fit(df)

  # Calcular el score
  score = round(k_means.score(df) * -1)

  # Mostrar el score
  print("K: " + str(k) + " - WSS: " + str(score))

  # Devolver el score
  return score

# Aplicar elalgoritmo para cada k
scores = [entrenar_kmeans(X, i) for i in k]
````

Para la selección del mejor K se sigue el **Método de Elbow**, este simplemente consiste en seleccionar el punto en el que agregar más clusters no resulta en mejores sustanciales en la métrica elegida. Para nuestro caso esa métrica es el WSS.

````python
# Graficar los resultados
plt.figure(figsize=PLOTSIZE)
plt.plot(k, scores, "bx-")
plt.xlabel("Número de Clusters")
plt.ylabel("WSS - Within Sum Squares")
plt.title("Curva Elbow")
plt.show()
````

Para este caso, la curva no resulta tener un "codo" bien definido. Esto quiere decir, en forma coloquial, que la curva decrece de una forma suave. 

Se decide tomar un valor de k = 10. Con este número se considera una buena reducción en el WSS, y es un número suficientemente chico con el que se puede sacar conclusiones acerca de los tipos de clientes de la empresa.
