## Descripción Datos

Los datos describen una encuesta de satisfacción llevada a cabo por una aerolínea estadounidense, se desconoce el nombre particular de la aerolínea. Algunos conceptos importantes para comprender el dominio del problema son:
* **Check-in**: es la confirmación de tu participación en el vuelo, se puede hacer tanto en forma presencial como online.
* **Pre-embarque**: es el proceso por el cual se despacha la valija, se recibe el boleto de embarque y se realizan algunos controles.
* **Embarque**: es el proceso de subida al avión, se realiza cerca del horario de despegue.
* **Inflight**: todas las actividades que se realizan mientras el avión se encuentra volando. 
* **Leg Room**: es la distancia que existe entre un punto de un asiento y el mismo punto en el asiento siguiente.

```python
# Importación de datos
URL_DATOS = "https://raw.githubusercontent.com/TomasSG/Parcial2-CienciaDeDatos/feature/eda/datos/data.csv"
COLUMN_SEP = ";"
DECIMAL_SEP = "."
```

## Importación de datos

Importamos los datos para comenzar a trabajar

```python
# Importamos los datos mediante el link
datos_raw = pd.read_csv(URL_DATOS, sep=COLUMN_SEP, decimal=DECIMAL_SEP)

# Verificamos que se hayan exportado correctamente
datos_raw.head(1)
```

## Análisis

Exploramos los datos aplicando las diferentes técnicas vistas.

El primer paso será, establecer la variable id como índice y modificameros el nombre de la variable target (satisfaction_v2) a "clase"

```python
# Hacemos una copia de los datos para trabajar
datos_eda = datos_raw.copy()

# Establecemos "id" como índice de las filas
datos_eda.set_index("id", inplace=True)

# Renombramos la variable target a "CLASE"
datos_eda["CLASE"] = datos_eda["satisfaction_v2"]
datos_eda.drop("satisfaction_v2", axis=1, inplace=True) # axis=1 para indicar que es una columna

# Verificamos que se hayan realizado correctamente las operaciones
datos_eda.head(1)
```

### Descripción Básica de los datos

```python
# Calculemos la cantidad de filas y columnas
print("El dataset continene {} filas".format(datos_eda.shape[0]))
print("El dataset continene {} columnas".format(datos_eda.shape[1]))
```

```python
# Tipo de columnas
print("Tipo de las columnas:")
print("-" * 50)
print(datos_eda.dtypes)
```

Observamos que hay varias calumnas que las trata como numéricas cuando en realidad representan variables categóricas

```python
# Definicmos las variables que vamos a cambiar su tipo
vars_to_objecto = ["Inflight wifi service", "Departure/Arrival time convenient", "Ease of Online booking", "Gate location", "Food and drink", "Online boarding",
                   "Seat comfort", "Inflight entertainment", "On-board service", "Leg room service", "Baggage handling", "Checkin service", "Inflight service",
                   "Cleanliness"]

# Realizamos los cambios
for var in vars_to_objecto:
  datos_eda[var] = datos_eda[var].astype("object")

# Verificamos que los tipos sean correctos ahora
print("Tipo de las columnas:")
print("-" * 50)
print(datos_eda.dtypes)
```
