# Preparación de datos

En esta sección comenzamos definiendo las funciones necesarias para aplicar la preparación de datos. Luego, generaremos diferentes datasets para realizar el entrenamiento de modelos y encontrar aquellos que tengan una mejor perfomance.

## Definición funciones
Con el objetivo de modularizar la función preparacion_datos, vamos a constuir algunas funciones que le sirven de soporte.

````python
# FUNCION imputar_nulos
# Para imputar los nulos en Arrival Delay in Minutes se decide reemplazar los valores por la media.
def imputar_nulos(df):

  # Copiar el df para no modificarlo in situ
  df_c = df.copy()

  # Calcular la media de las observaciones, sin tener en cuenta los nulos
  index = pd.notnull(df_c["Arrival Delay in Minutes"])
  media = df_c[index]["Arrival Delay in Minutes"].mean()
  
  # Reemplazar los valores
  df_c["Arrival Delay in Minutes"] = df_c["Arrival Delay in Minutes"].fillna(media)

  return df_c
````

````python
# FUNCION calcular_media_por_derecha
# Calcula la media de la variable indicada, sin tener encuenta los valores que sean mayores al maximo dado
def calcular_media_por_derecha(df, var, max):
  
  # Filtrar los datos mayores a max
  index = df[var] <= max
  df_aux = df[index]

  # Calcular la media con los datos filtrados
  media = df_aux[var].mean()

  return media

# FUNCION imputar_outliers
# Para imputar los outliers en las variables necesarias, siempre por derecha.
def imputar_outliers(df):

  # Copiar el df para no modificarlo in situ
  df_c = df.copy()

  # Definir variables a las que aplicar la imputacion
  vars = ["Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]

  for var in vars:

    # Definimos cual será el outlier superior
    outlier_sup = df_c[var].mean() + 2 * df_c[var].std()

    # Calculamos la media sin valores mayores al outlier
    media_sin_outlier = calcular_media_por_derecha(df_c, var, outlier_sup)

    # Imputamos los valores
    df_c[var] = [x if x <= outlier_sup else media_sin_outlier for x in df_c[var]]

  return df_c
````

````python
# FUNCION crear_variables
# Para aplicar feature engineering.
def crear_variables(df):

  # Copiar el df para no modificarlo in situ
  df_c = df.copy()

  # 1) Variable que representa el delay total en un viaje
  df_c["Total Delay in Minutes"] = df_c["Departure Delay in Minutes"] + df_c["Arrival Delay in Minutes"]

  # 2) Diferencia entre los delay, para conocer cómo varía
  df_c["Diff Delay in Minutes DA"] = df_c["Departure Delay in Minutes"] - df_c["Arrival Delay in Minutes"]

  return df_c
````

````python
# FUNCION preparacion_datos
# df                ->  dataset de entrada a modifiar
# flag_outliers  ->  flag para imputar outliers
# flag_nulos     ->  flag para imputar nulos
# flag_vars        ->  flag para feature engineering
def preparacion_datos(df, flag_outliers, flag_nulos, flag_vars):

  # Crear una copia
  df_c = df.copy()

  # ---- IMPUTACIÓN DE OUTLIERS ----
  if flag_outliers:
    df_c = imputar_outliers(df_c)

  # ---- IMPUTACIÓN NULOS ----
  if flag_nulos:
    df_c = imputar_nulos(df_c)

  # ---- FEATURE ENGINEERING ----
  if flag_vars:
    df_c = crear_variables(df_c)

  # ---- TARGET ----
  # Converitr target en numerica
  # 0 ->  neutral or dissatisfied 
  # 1 ->  satisfied
  df_c["CLASE"] = [1 if x == "satisfied" else 0 for x in df_c["CLASE"]]
  y = df_c["CLASE"]

  # ---- FEATURES ----
  X = df_c.drop("CLASE", axis=1)

  # Normalizar las numéricas
  num_cols = X.columns[X.dtypes.apply(lambda c: np.issubdtype(c, np.number))]
  scaler_X = StandardScaler(with_mean=True, with_std=True)
  X[num_cols] = scaler_X.fit_transform(X[num_cols])

  # Obtenemos las dummies
  X = pd.get_dummies(X)

  return y, X
````

## Procesar Datos

En esta sección vamos procesar los datos, usando las función definida previamente.

Elegimos crear dos conjuntos de datos diferentes:

1. Un dataset aplicando todas las transformaciones.
2. Un datset sin imputar outliers.

````python
# Generamos datos aplicando todas las transformaciones
y, X_todas = preparacion_datos(datos_eda, flag_outliers=True, flag_nulos=True, flag_vars=True)

# Generamos datos sin imputar outliers
_, X_con_outliers = preparacion_datos(datos_eda, flag_outliers=False, flag_nulos=True, flag_vars=True)
````

## División Datos
Vamos a dividir los datos en test y train,  usamos la  siguiente función

```
train_test_split(X, y, test_size, random_state)
```

Se elige una proporción de 30/70, es decir, 30% para test y 70% para train.

Para que el trabajo sea reproducible se va a utilizar una seed, tendrá el valor de 150.


````python
# Usando el conjunto de datos con todas las transformaciones aplicadas
X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(X_todas,
                                                            y,
                                                            test_size=TEST_SIZE,
                                                            random_state=RANDOM_STATE)

# Usando el conjunto de datos con outliers
X_train_o, X_test_o, y_train_o, y_test_o = train_test_split(X_con_outliers,
                                                            y,
                                                            test_size=TEST_SIZE,
                                                            random_state=RANDOM_STATE)
````
