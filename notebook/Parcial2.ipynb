# Contexto de Datos
## Plataforma
Los datos fueron extraidos de la plataforma [Kaggle](https://www.kaggle.com/). Esta es una plataforma centrada en las temáticas de ciencia de datos y machine learning. Aquí, los miembros pueden publicar set de datos abiertos, subir notebooks que ellos desarrollan, escribir preguntas, etc.

Además, se desarrollan competencias para que la comunidad pueda participar y resolver problemas. Las competencias van desde simples ejercicios para aprender conceptos básicos, hasta desafíos propuestos por grandes empresas, donde los ganadores reciben enorme sumas de dinero.

## Descripción Datos
**Fuente**: [link](https://www.kaggle.com/johndddddd/customer-satisfaction)
