## Descripción Datos
**Fuente**: [link](https://www.kaggle.com/johndddddd/customer-satisfaction)

Los datos describen una encuesta de satisfacción llevada a cabo por una aerolínea estadounidense, se desconoce el nombre particular de la aerolínea. Algunos conceptos importantes para comprender el dominio del problema son:
* **Check-in**: es la confirmación de tu participación en el vuelo, se puede hacer tanto en forma presencial como online.
* **Pre-embarque**: es el proceso por el cual se despacha la valija, se recibe el boleto de embarque y se realizan algunos controles.
* **Embarque**: es el proceso de subida al avión, se realiza cerca del horario de despegue.
* **Inflight**: todas las actividades que se realizan mientras el avión se encuentra volando. 
* **Leg Room**: es la distancia que existe entre un punto de un asiento y el mismo punto en el asiento siguiente.

```python
# Hacemos una copia de los datos para trabajar
datos_eda = datos_raw.copy()

# Establecemos "id" como índice de las filas
datos_eda.set_index("id", inplace=True)

# Renombramos la variable target a "CLASE"
datos_eda["CLASE"] = datos_eda["satisfaction_v2"]
datos_eda.drop("satisfaction_v2", axis=1, inplace=True) # axis=1 para indicar que es una columna

# Verificamos que se hayan realizado correctamente las operaciones
datos_eda.head(1)
```

### Análisis variables numéricas
Vamos a observar el comportamiento de las variables numéricas numéricas, en búsqueda de outliers.

```python
# Definimos cuales son las variables numéricas
vars_num = ["Age", "Flight Distance", "Departure Delay in Minutes", "Arrival Delay in Minutes"]

# Graficamos los boxplot en busca de outliers
for var in vars_num:
  plt.figure(figsize=FIGSIZE)
  sns.boxplot(x=var, data=datos_eda, orient="h")
```

Observamos que:

* **Flight distance** presenta unos pocos outliers a derecha.
* **Departure Delay** presenta muchos outliers a derecha.
